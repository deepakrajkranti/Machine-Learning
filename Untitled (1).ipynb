```python
data=df.values
print(df.shape)
```

```python
x=data[:,1:]
y=data[:,0]
print(x.shape,y.shape)
```

```python
split=int(0.8*x.shape[0])
```

```python
x_train=x[:split,:]
y_train=y[:split]

x_test=x[split:,:]
y_test=y[split:]
print(x_test,y_test)
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(x,y,querypoint,k=5):
    
    vals=[]
    m=x.shape[0]
    
    for i in range(m):
        d=dist(querypoint,x[i])
        vals.append((d,y[i]))
        
        
    vals=sorted(vals)
    vals[:k]
    
    vals=np.array(vals)
    
    new_vals=np.unique(vals[:,1],return_counts=True)
    
    index=new_vals[1].argmax()
    
    pred=new_vals[0][index]
    
    return pred
    
```

```python
def draw(sample):
    img=sample.reshape((28,28))
    plt.imshow(img,cmap='gray')
    plt.show()
    
draw(x_train[3])
```
